### Formatação dos dados para um formato adequeado para a função de visualização

```python
loops = dict()
loop_list = {}
workspaces = {}
workspaces_list = []

for conversation in data:
    nodes = {}
    workspace_id = conversation[0]['workspace_id']
    for i in range(len(conversation)):
        
        try:
            nodes[conversation[i]['node_id']] += 1
        except:
            nodes[conversation[i]['node_id']] = 1

    for node in nodes:
        try:
            loops[workspace_id][nodes[node]][node] += 1
        except:
            try:
                loops[workspace_id][nodes[node]][node] = 1
            except:
                try:
                    loops[workspace_id][nodes[node]] = {}
                    loops[workspace_id][nodes[node]][node] = 1
                except:
                    loops[workspace_id] = {}
                    loops[workspace_id][nodes[node]] = {}
                    loops[workspace_id][nodes[node]][node] = 1
```

```python
for workspace in loops:
    loop_list[workspace] = {}
    workspaces[workspace] = 0
    for num in loops[workspace]:
        loop_list[workspace][num] = []
        for node in loops[workspace][num]:
            loop_list[workspace][num].append({'node_id': node, 'freq': loops[workspace][num][node]})
            workspaces[workspace] += loops[workspace][num][node]
            
for workspace in workspaces:
    workspaces_list.append({'workspace': workspace, 'loops': workspaces[workspace]})
```
