## Tratamento de dados faltantes

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/aluguel2.csv',sep=';')
```

```python
df
```

```python
na = df.isnull()
na
```

```python
df.notnull()
```

```python
#eu consigo ter uma visão sobre quais atributos tem valores nulos. Onde na função .info().
#Nele é exibido a qtd de registros dos atributos que não são nulos

df.info()
```

```python
#Pegar valores que são nulos 
df[df['Valor'].isnull()]
```

```python
antes_drop = len(df)
df.dropna(subset=['Valor'],inplace=True)
depois_drop = len(df)
antes_drop - depois_drop
```

```python
df
```

```python
df[df['Valor'].isnull()]
```

```python
#No desafio dado no curso, a regra de negócio é que eu preciso tirar todo o apartamento precisa ter condominio
df[df['Condominio'].isnull()].shape[0]
```

```python
selecao = (df['Tipo']=='Apartamento')&(df['Condominio'].isnull())
```

```python
antes_drop = len(df)
df = df[~selecao]
depois_drop = len(df)
antes_drop - depois_drop
```

```python
df[df['Condominio'].isnull()].shape[0]
```

```python
#Atribuir valores em Nan
df.fillna(0, inplace = True)
#Ou poderíamos substituir pela média: s.fillna(s.mean())
```

```python
df.info()
```

```python
df.to_csv('data/aluguel2.csv',sep=';',index=False)
```

## Identificando e removendo outliers

```python
df.boxplot('Valor')
```

```python
df[df['Valor']>=500000]
```

```python
dados = df['Valor']
```

```python
Q1 = dados.quantile(.25)
```

```python
Q1
```

```python
Q3 = dados.quantile(.75)
```

```python
amp = Q3 - Q1
limite_inferior = Q1 - 1.5 * amp
limite_superior = Q3 + 1.5 * amp
```

```python
selecao = (dados >= limite_inferior) &(dados <= limite_superior)
dados_new = df[selecao]
dados_new.boxplot('Valor')
```

```python
df.hist('Valor')
dados_new.hist('Valor')
```
